# **One-Hot Encoding**

```python
from sklearn.preprocessing import OneHotEncoder

# Identify categorical columns
categorical_cols = data.select_dtypes(include=['object']).columns

# Apply One-Hot Encoding
data_encoded = pd.get_dummies(data, columns=categorical_cols, drop_first=True)

# Now apply Standardization
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_encoded)
```

```python
#Anjali Srivastava
#21BDS0069
# Define target variable (Assuming last column is the target)
target_column = data['headache']
X = data.drop('headache',axis=1)
```

```python
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder
# Apply Label Encoding for categorical columns
categorical_cols = X.select_dtypes(include=['object']).columns
le = LabelEncoder()
for col in categorical_cols:
    X[col] = le.fit_transform(X[col])
```

```python
le.fit_transform(target_column)
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X,target_column, test_size=0.2, random_state=42)
```

```python
y_test.shape
#Anjali Srivastava
#21BDS0069
```

```python
X.head()
#Anjali Srivastava
#21BDS0069
```

```python
y_train = le.fit_transform(y_train)
#Anjali Srivastava
#21BDS0069
```

```python
y_test = le.fit_transform(y_test)
#Anjali Srivastava
#21BDS0069
```

# **Model Building**

```python
# Build Simple Linear Regression Model
model = LinearRegression()
model.fit(X_train, y_train)
#Anjali Srivastava
#21BDS0069
```

# **Prdeiction**

```python
# Predictions
y_pred = model.predict(X_test)
#Anjali Srivastava
#21BDS0069
```

```python
#Before that, let's predict the y values with our model and evaluate it more. And a target variables DataFrame is also built:
# a data frame with actual and predicted values of y
evaluate = pd.DataFrame({'Actual': y_test.flatten(),
'Predicted': y_pred.flatten()})
evaluate.head(10)
#Anjali Srivastava
#21BDS0069
```

```python
#The preceding screenshot shows the difference between the actual values and the predicted values. We can see them if we plot them:
evaluate.head(10).plot(kind = 'bar')
#Anjali Srivastava
#21BDS0069
```

# **Evaluation**

```python
#Computing accuracy
#Anjali Srivastava
#21BDS0069
# Scoring the model
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
# R2 Score
print(f"R2 score: {r2_score(y_test, y_pred)}")
# Mean Absolute Error (MAE)
print(f"MAE score: {mean_absolute_error(y_test, y_pred)}")
# Mean Squared Error (MSE)
print(f"MSE score: {mean_squared_error(y_test, y_pred)}")
```

# **Some other models**

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
```

```python
# Define models
#Anjali Srivastava
#21BDS0069
models = {
    "Linear Regression": LinearRegression(),# linear regression model
    "Decision Tree": DecisionTreeRegressor(random_state=42), #decision tree
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42), #Random Forest model
    "Support Vector Regressor": SVR(kernel='rbf'), #SVM model
    "K-Nearest Neighbors": KNeighborsRegressor(n_neighbors=5),#K -nearest neighbor
    "Gradient Boosting": GradientBoostingRegressor(n_estimators=100, random_state=42), # gradient boost
    "XGBoost": XGBRegressor(objective='reg:squarederror', n_estimators=100, random_state=42) #xgboost
}
```

# **Prediction and Evaluation**

```python
#Anjali Srivastava
#21BDS0069
# Train and evaluate each model
results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    results[name] = {
        "R2 Score": r2_score(y_test, y_pred),
        "MAE": mean_absolute_error(y_test, y_pred),
        "MSE": mean_squared_error(y_test, y_pred)
    }
```

```python
 #Anjali Srivastava
#21BDS0069
 # Display results
results_df = pd.DataFrame(results).T
print(results_df.sort_values(by='R2 Score', ascending=False))
```
